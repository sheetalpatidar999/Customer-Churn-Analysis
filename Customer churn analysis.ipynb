```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("D:/data analyst projects/Python/projects/customer Churn.csv")
df.head()
```

#REPLACE BLANKS WITH ZERO AND CONVERT DATA TYPE FROM STRING TO FLOAT

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype(float)
```

#convert 0 and 1 into yes /no in seniorcitizen column

```python
def conv(value):
    if value ==1:
        return'yes'
    else:
        return'no'
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
plt.figure(figsize=(3,4))
ax=sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0])
plt.title('Count of Coustomer by Churn')
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct='%1.2f%%')
plt.title('Percentage of Chunred Customer')
plt.show()
```

#from given pie chart we can conclude that 26.54% of our customer churned out.

```python
import pandas as pd
import matplotlib.pyplot as plt

# STEP 1: Group and calculate percentage
group = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
group_percent = group.div(group.sum(axis=1), axis=0) * 100

# STEP 2: Plotting stacked bar chart
ax = group_percent.plot(kind='bar', stacked=True, 
                        figsize=(5,5))

# STEP 3: Add percentage labels
for i, row in enumerate(group_percent.values):
    bottom = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, bottom + val/2, f'{val:.1f}%', ha='center', va='center', fontsize=9)
            bottom += val

# STEP 4: Final touches
plt.title("Churn Percentage by Senior Citizen")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.xticks(ticks=[0,1], labels=['No','Yes'], rotation=0)
plt.legend(title="Churn", loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

#here comparatively a big percentege of senior citizen category has churned out.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Column list
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Grid size (3 rows x 3 columns for 9 plots)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(16, 12))
axes = axes.flatten()  # 2D array ko flat 1D bana do for easy looping

# Plotting each subplot
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], hue='Churn')
    axes[i].set_title(f"{col} Count")
    axes[i].tick_params(axis='x', rotation=360)  # Rotate labels if long
    for container in axes[i].containers:
        axes[i].bar_label(container, fontsize=8)  # Show count on bars

# Layout adjustment
plt.tight_layout()
plt.suptitle("Service-wise Customer Distribution", fontsize=18, y=1.03)
plt.show()
```

#Customers without services like OnlineSecurity, TechSupport, DeviceProtection, OnlineBackup, and Streaming are more likely to leave (churn).
People using Fiber optic internet also have higher churn than those using DSL.
Having PhoneService or MultipleLines doesn’t affect churn much, but overall, more active services = less chance of leaving.
